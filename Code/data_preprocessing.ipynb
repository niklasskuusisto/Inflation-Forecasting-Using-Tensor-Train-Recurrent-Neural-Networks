```python
# file input and output paths
input_path = '/content/drive/MyDrive/Colab Notebooks/Gradu/Data/Excel preprocessed/'
output_path = '/content/drive/MyDrive/Colab Notebooks/Gradu/Data/Python preprocessed/'
```

```python
financial_variables = ['oil', 'silver', 'eurusd', 'eurcfh', 'spx', 'omx']
macro_variables = ['inflation', 'unemployment', 'imports', 'exports', 'exchange', 'ppi', 'bci', 'cci', 'construction', 'manufacturing', 'world', 'share_prices']
```

```python
countries = ['Belgium', 'Denmark', 'Germany', 'Spain', 'France', 'Italy', 'Netherlands', 'Austria', 'Poland', 'Finland', 'Sweden', 'United States']
```

```python
for variable in financial_variables:
    input_file = input_path + variable + '.csv'
    output_file = output_path + variable + '.csv'

    data = pd.read_csv(input_file, header=0, index_col=0)
    data = data.astype(float) # set to float type
    data = data.drop(columns=['Open', 'High', 'Low', 'Vol.', 'Change ']) # drop columns
    data = data[::-1]
    data.index = pd.to_datetime(data.index, format = '%m/%d/%Y').strftime('%Y-%m')

    df_all = pd.DataFrame(index=data.index) # initialise with index in order to copy data over
    df_all = df_all.astype(float)
    for country in countries:
        df_all[country] = data["Price"]
    df_all.to_csv(output_file) # save to csv

    print(variable, df_all)
```

```python
for variable in macro_variables:
    input_file = input_path + variable + '.csv'
    data = pd.read_csv(input_file, header=0, index_col=0)
    data = data.astype(float) # set to float type
    data = data.drop(columns=['Ireland'])
    output_file = output_path + variable + '.csv'
    data.to_csv(output_file)
    print(variable, data)
```
